```python
import qc
import numpy as np
from qutip import *
from qc.bloch_matrix import *
```

```python
b = qutip.Bloch()
b.make_sphere()
```

```python
b.clear()
n = 100
eps = 1
rz = get_rotation_matrix(Axis.Z, eps)
ry = get_rotation_matrix(Axis.Y, eps)
v = [1/np.sqrt(2), 1/np.sqrt(2), 0]
pnts = [v]

for i in range(n):
    pnts.append(rz.dot(pnts[i]))
    pnts.append(ry.dot(pnts[i]))

b.add_points(np.transpose(pnts))
b.render()
```
